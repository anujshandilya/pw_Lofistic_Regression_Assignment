1. **What is Logistic Regression, and how does it differ from Linear Regression?**
- Logistic Regression is a statistical model used for binary classification, meaning it predicts outcomes that belong to one of two categories (e.g., Yes/No, Pass/Fail, 1/0). It differs from Linear Regression in several key ways:
  - Linear Regression predicts continuous values (e.g., house prices, salaries).
  - Logistic Regression predicts categorical outcomes (e.g., whether a customer will buy a product).
  - Linear Regression gives raw numeric predictions.
  - Logistic Regression outputs probabilities, then classifies based on a threshold (e.g., 0.5 for binary classification).
  - Linear Regression models relationships between numeric variables.
  - Logistic Regression is best for classification tasks like spam detection, fraud detection, and medical diagnosis.

5. **What is Regularization in Logistic Regression? Why is it needed?**
- Regularization in Logistic Regression helps prevent overfitting by adding a penalty term to the loss function, ensuring the model generalizes well to new data.
- It is needed because:
  - When the model learns too much from training data, it loses generalization ability.
  - Limits large coefficients that might overemphasize certain features.
  - Helps stabilize models when predictors are highly correlated.

6. **Explain the difference between Lasso, Ridge, and Elastic Net regression.**
- Ridge Regression (L2 Regularization):
  - Adds a penalty to large coefficients by using the L2 norm.
  - Shrinks coefficients but doesn’t set them to zero → all features remain in the model.
  - Best for Handling multicollinearity, when predictors are highly correlated.

- Lasso Regression (L1 Regularization):
  - Uses the L1 norm penalty:
  - Shrinks some coefficients to exactly zero, effectively performing feature selection.
  - Best for Sparse models where only the most important predictors should be retained.

- Elastic Net Regression (L1 + L2 Regularization):
  - Combines Ridge and Lasso penalties:
  - Balances both feature selection (like Lasso) and coefficient shrinkage (like Ridge).
  - Best for Cases where correlated features exist, and feature selection is needed.

7. **When should we use Elastic Net instead of Lasso or Ridge?**
- Elastic Net is preferred over Lasso or Ridge when:
  - If predictors are highly correlated, Ridge helps stabilize coefficients, while Lasso can eliminate variables. Elastic Net balances both effects.
  - Lasso performs aggressive feature selection by shrinking some coefficients to zero, but Elastic Net selects while retaining some influence from correlated features.
  - In datasets with many features, Elastic Net helps prevent overfitting, ensuring both sparsity (Lasso) and regularization (Ridge) work together.
  - Lasso can randomly pick a single variable when features are correlated, while Elastic Net preserves stability by distributing weights.
  - Best Use Case → When the dataset has many predictors, correlations among variables, and a need for both regularization and feature selection.

8. **What is the impact of the regularization parameter (A) in Logistic Regression?**
- Logistic Regression, the regularization parameter plays a crucial role in controlling the complexity of the model and preventing overfitting.
- Impact of Regularization Parameter:
  - Higher Regularization → Stronger penalty, forcing coefficients to shrink toward zero.
  Reduces model complexity, but excessive regularization can cause underfitting, leading to poor predictions.
  - Lower Regularization → Weaker penalty, allowing coefficients to take larger values.
  Helps capture more patterns in data, but too little regularization may cause overfitting, making the model overly sensitive to noise.

14. **What are different solvers in Logistic Regression? Which one should be used?**
- In Logistic Regression, different solvers optimize the model’s parameters efficiently. The choice of solver depends on dataset size, regularization type, and computational efficiency.
- Common Solvers in scikit-learn's Logistic Regression
  - lbfgs → Best for L2 regularization, works well with large datasets, and is default for multiclass problems.
  - liblinear → Supports both L1 & L2 regularization, better for small datasets or sparse data.
  - saga → Ideal for L1, L2, and Elastic Net regularization, suitable for large datasets and online learning.
  - newton-cg → Works well with L2 regularization, useful for multiclass classification but computationally expensive.
  - sag → Good for large datasets, optimized for L2 regularization, but requires normalized features.
- Which Solver Should we Use?
  - Use lbfgs for standard logistic regression with L2 regularization.
  - Use liblinear for small datasets and L1 regularization (feature selection).
  - Use saga when working with large datasets and Elastic Net regularization.
  - Use newton-cg for multiclass classification if computational cost isn't a concern.
  - Use sag for very large datasets, but ensure features are scaled for stability.

15. **How is Logistic Regression extended for multiclass classification?**
- Logistic Regression is naturally designed for binary classification, but it can be extended to multiclass classification using different techniques.
1. One-vs-Rest (OvR) Approach
  - Also called One-vs-All (OvA)
  - Trains multiple binary classifiers, each distinguishing one class vs. all others
  - For N classes, Logistic Regression runs N separate models
  - At prediction time, the class with the highest probability is selected
  - Works well but can struggle with highly imbalanced classes
2. Softmax Regression (Multinomial Logistic Regression)
  - Instead of multiple binary classifiers, a single model predicts all classes at once
  - Uses the Softmax function to convert logits into probabilities.
  - Produces a probability distribution across all classes, ensuring they sum to 1
  - More efficient than OvR when classes are highly related.

16. **What are the advantages and disadvantages of Logistic Regression?**
- Advantages:
  - Simple & Interpretable → Easy to implement and interpret, making it a great starting point for classification tasks.
  - Efficient for Small Datasets → Works well even when the dataset size is limited.
  - Probability-Based Predictions → Outputs probabilities, making decision thresholds adjustable.
  - Handles Linearly Separable Data Well → When data follows a linear boundary, Logistic Regression performs efficiently.
  - Regularization Support → Can incorporate L1 (Lasso) and L2 (Ridge) regularization to prevent overfitting.
- Disadvantages:
  - Assumes Linear Decision Boundary → Struggles with complex patterns in  non-linearly separable data.
  - Sensitive to Outliers → Extreme values can distort the regression coefficients significantly.
  - Requires Feature Scaling → Works best when features are normalized (e.g., using StandardScaler).
  - Less Effective for Large Feature Sets → High-dimensional data may require more complex models (e.g., SVM, neural networks).
  - Handles Multicollinearity Poorly → Highly correlated predictors can make the model unstable; Ridge regularization helps mitigate this.

  - Uses the sigmoid function to compute probabilities.

19. **How do we choose between One-vs-Rest (OVR) and Softmax for multiclass classification?**
- One-vs-Rest (OvR)
  - How it works: Trains N separate binary classifiers, each distinguishing one class vs. all others.
  -  Best for:
  - Sparse datasets (where many feature values are zero).
  - Interpretable models (easier to analyze individual class decisions).
  - Small datasets, as it avoids complexity of handling probabilities across all classes.

- Softmax Regression
  - How it works: A single model predicts all classes at once using Softmax function:
  - Best for:
    - Balanced datasets, where all classes have sufficient instances.
    - Multiclass problems with interrelated categories (e.g., digit classification).
    - More computationally efficient, as it avoids training multiple classifiers.

25. Write a Python program to train Logistic Regression, save the trained model using joblib, and load it again to

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_iris
dataset = load_iris()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['target'] = dataset.target
df = df[df['target'] != 2]

# Seperating x and y
x=df.drop('target', axis = 1)
y=df['target']

# Train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=1)

#Model Training
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)

# Saving the model
import joblib
joblib.dump(lr, "Logistic_model.pkl")
print("Saved")

# Load the model and predict
load_lr = joblib.load("Logistic_model.pkl")
print("Loaded")
ypred = load_lr.predict(x_test)

print(ypred)
```
